```python
def quality(total_bill,tip):
    if tip/total_bill>0.25:
        return "generous"
    else:
        return "Other"
```

```python
quality(16.99,1.01)
```

```python
#method 1
df['Quality1']=df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
```

```python
#method 2 faster
df['Quality2']=np.vectorize(quality)(df['total_bill'],df['tip'])
```

```python
setup='''
import pandas as pd
import numpy as np

df=pd.read_csv('C:\\Users\\kunal\\OneDrive\\CODE FOR LIFE\\Python projects\\Python Libraries\\Untitled Folder\\tips.csv')#csv_file-> comma seperated file
def quality(total_bill,tip):
    if tip/total_bill>0.25:
        return "generous"
    else:
        return "Other"

'''
```
