```python
#CSV
import pandas as pd
from io import StringIO,BytesIO
data=('col1;col2;col3\n'
     'a;b;c\n'
     'x;y;z\n'
     'm;n;o\n')

print(data)
```

```python
type(data)
```

```python
StringIO()
```

```python
#CSV=pd.read_csv(data,sep=';') error
CSV=pd.read_csv(StringIO(data),sep=';')
```

```python
CSV
```

```python
CSV=pd.read_csv(StringIO(data),sep=';',usecols=['col2'])
CSV
```

```python
CSV=pd.read_csv(StringIO(data),sep=';',dtype=object)
CSV
```

```python
CSV=pd.read_csv(StringIO(data),sep=';',dtype=int)
CSV
```

```python
CSV=pd.read_csv(StringIO(data),sep=';',dtype=character)
CSV
```

```python
CSV['col1'][2]
```

```python
CSV=pd.read_csv(StringIO(data),dtype={
    'col1':int,
    'col2':float,
    'col3':'Int64'
})
CSV
```
