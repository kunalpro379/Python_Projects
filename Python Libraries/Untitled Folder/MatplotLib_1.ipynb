```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Create subplots
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r--')

plt.subplot(2, 2, 2)
plt.plot(x, y, 'g*-')

plt.subplot(3, 2, 3)  
plt.plot(x, y, 'bo')

plt.subplot(3, 2, 4)
plt.plot(x, y, 'm-')

plt.subplot(3, 2, 5)
plt.plot(x, y, 'c--')

plt.subplot(3, 2, 6)
plt.plot(x, y, 'y*-')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 4*np.pi, 0.1)
y = x**2 * (np.sin(2*x**2 - 6*x + 99)) - x * (np.cos(9*x**2 * 99 - 9)) + 102

plt.title("Some random function waveform")
plt.plot(x, y,'ro-')
plt.show()
```
